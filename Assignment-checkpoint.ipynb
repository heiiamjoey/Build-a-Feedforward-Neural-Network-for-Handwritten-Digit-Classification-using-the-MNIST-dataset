```python
# Import libraries
import numpy as np  # Numerical operations
import idx2numpy # convert .idx files into numpy files
from tensorflow.keras.models import Sequential  # create a linear stack of layers
from tensorflow.keras.layers import Dense, Flatten #layer for the neural network
from tensorflow.keras.utils import to_categorical # convert class labels into one-hot encoded format
from sklearn.model_selection import train_test_split # split dataset into training and testing stes
from sklearn.metrics import confusion_matrix,classification_report  # computer confusion matrix to evaluate classification results
import seaborn as sns  # visualizing the confusion matrix
import matplotlib.pyplot as plt # plotting library
```

```python
# Load the dataset
X_train = idx2numpy.convert_from_file('x_train.idx3-ubyte')
y_train = idx2numpy.convert_from_file('y_train.idx1-ubyte')
X_test = idx2numpy.convert_from_file('x_test.idx3-ubyte')
y_test = idx2numpy.convert_from_file('y_test.idx1-ubyte')

print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")
```

```python
# Preprocess the data
# Normalize (scale pixel values to [0, 1])
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
# Split into training and validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
```

```python
# One-hot encode labels
y_train_cat = to_categorical(y_train, 10)
y_val_cat = to_categorical(y_val, 10)
y_test_cat = to_categorical(y_test, 10)
```

```python
# Function to build the Neural Network Model
def build_model(hidden_units):
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(hidden_units, activation='relu'),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
# Define training configurations
configs = [
    {'name': 'Config 1', 'hidden_units': 64, 'epochs': 10, 'batch_size': 64},
    {'name': 'Config 2', 'hidden_units': 128, 'epochs': 10, 'batch_size': 128},
    {'name': 'Config 3', 'hidden_units': 128, 'epochs': 15, 'batch_size': 64}
]

histories = []
models = []
```

```python
# Train all configurations and store their histories and models
for config in configs:
    print(f"\n=== Training {config['name']}: {config['hidden_units']} units, "
          f"{config['epochs']} epochs, batch size {config['batch_size']} ===")
    model = build_model(config['hidden_units'])
    history = model.fit(
        X_train, y_train_cat,
        epochs=config['epochs'],
        batch_size=config['batch_size'],
        validation_data=(X_val, y_val_cat),
        verbose=2
    )
    val_acc = history.history['val_accuracy'][-1]
    print(f"{config['name']} - Final Validation Accuracy: {val_acc:.4f}")
    histories.append(history)
    models.append(model)
```

```python
# Evaluate all trained models on test data
print("\nTest accuracy for each configuration:")
test_accuracies = []
for model, config in zip(models, configs):
    test_loss, test_acc = model.evaluate(X_test, y_test_cat, verbose=0)
    print(f"{config['name']} Test Accuracy: {test_acc:.4f}")
    test_accuracies.append(test_acc)
```

```python
# Select the best model (Config 3) for detailed evaluation and plotting
best_model = models[-1]  # Config 3 is the best based on the highest accuracy
```

```python
# Predict test labels with best model
y_pred_probs = best_model.predict(X_test)
y_pred_classes = np.argmax(y_pred_probs, axis=1)
```

```python
# Plot training & validation loss/accuracy for each configuration
for history, config in zip(histories, configs):
    epochs_range = range(1, len(history.history['loss']) + 1)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, history.history['loss'], 'b-', label='Training Loss')
    plt.plot(epochs_range, history.history['val_loss'], 'r-', label='Validation Loss')
    plt.title(f"{config['name']} - Loss")
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, history.history['accuracy'], 'b-', label='Training Accuracy')
    plt.plot(epochs_range, history.history['val_accuracy'], 'r-', label='Validation Accuracy')
    plt.title(f"{config['name']} - Accuracy")
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# Plot confusion matrix
conf_mat = confusion_matrix(y_test, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Best Model (Config 3)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# Print classification report for best model
print("Classification Report - Best Model (Config 3):")
print(classification_report(y_test, y_pred_classes))
```

```python
# Show sample predictions with images for best model
fig, axes = plt.subplots(3, 3, figsize=(8, 8))
axes = axes.flatten()
for i, ax in enumerate(axes):
    ax.imshow(X_test[i], cmap='gray')
    ax.set_title(f"True: {y_test[i]}\nPred: {y_pred_classes[i]}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```
